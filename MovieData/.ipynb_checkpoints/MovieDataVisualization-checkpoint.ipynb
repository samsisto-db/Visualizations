```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
#Read users dataset into a dataframe and assign header names
users = pd.read_table('u.user', sep='|', header=None, names=['index', 'age', 'sex', 'occupation', 'zipcode'], index_col='index')
```

```python
#Examine the users dataframe
users.head()
users.tail()
users.shape
```

```python
#Examine a histogram of user ages
sns.distplot(users.age)
```

```python
#Examine count of users by gender
viz1 = sns.countplot(x='sex', data=users)
```

```python
#Examine occupations in descending order of popularity
viz2 = sns.countplot(y='occupation', data=users, order=users.occupation.value_counts().index)
#viz2.set_xticklabels(viz2.get_xticklabels(), rotation=90)
```

```python
#Examine distribution of ages by gender
#sns.boxplot(x='sex', y='age', data=users)

#Examine distribution of ages by occupation in descending order of median value
x = users.groupby('occupation').age.median()
x = x.sort_values(ascending=False)
viz3 = sns.boxplot(x='occupation', y='age', data=users, order=x.index)
viz3.set_xticklabels(viz3.get_xticklabels(), rotation=90)
viz3
```

```python
#Examine distribution of ages by occupation and by sex
viz4 = sns.boxplot(x='occupation', y='age', hue='sex', data=users)
viz4.set_xticklabels(viz4.get_xticklabels(), rotation=90)
viz4
```

```python
#Examine the counts of female versus male doctors and lawyers as well as the mean age for each occupation
docs_and_lawyers = users[users.occupation.isin(['doctor', 'lawyer'])]
docs_and_lawyers.sex.value_counts()
docs_and_lawyers.groupby('occupation').age.mean()
```

```python
#Read ratings dataset into a dataframe and assign header names
ratings = pd.read_table('u.data', sep='\t', header=None, names=['user_id', 'movie_id', 'rating', 'timestamp'])
movie_info = pd.read_table('u.item', sep='|', header=None, names=['movie_id', 'movie_title', 'release_date', 'video_release_date',
              'IMDb_URL', 'unknown' , 'Action' , 'Adventure' , 'Animation' ,
              'Childrens' , 'Comedy' , 'Crime' , 'Documentary' , 'Drama' , 'Fantasy' ,
              'FilmNoir' , 'Horror' , 'Musical' , 'Mystery' , 'Romance' , 'SciFi' , 'Thriller' , 'War' , 'Western'], encoding = "ISO-8859-1")
```

```python
rating_w_names = ratings.join(movie_info, on='movie_id', how='left', rsuffix='_info')
```

```python
#rating_w_names.groupby('movie_title').rating.mean().sort_values(ascending=False)
#rating_w_names.movie_title.value_counts()
```

```python
rating_w_names['Year'] = rating_w_names.release_date.str[7:]
rating_w_names.Year = pd.to_numeric(rating_w_names.Year, errors='coerce').fillna(0).astype(np.int64)
```

```python
viz5 = sns.regplot(x='Year', y='rating', data=rating_w_names)
```

```python
#Do certain genres get higher ratings than others?

list_of_genres = ['Action' , 'Adventure' , 'Animation' ,
              'Childrens' , 'Comedy' , 'Crime' , 'Documentary' , 'Drama' , 'Fantasy' ,
              'FilmNoir' , 'Horror' , 'Musical' , 'Mystery' , 'Romance' , 'SciFi' , 'Thriller' , 'War' , 'Western']

y=[]

for genre in list_of_genres:
    print(genre)
    print(rating_w_names[rating_w_names[genre]==1].rating.mean())
    print(len(rating_w_names[rating_w_names[genre]==1]))
    y.append(len(rating_w_names[rating_w_names[genre]==1]))
```

```python
#Examine count of movies by genre
fig, ax = plt.subplots()    
width = 0.75
ind = np.arange(len(y))
ax.barh(ind, y, width, color="blue")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(list_of_genres, minor=False)
plt.title('Count of Movies by Genre')
plt.xlabel('Count')
plt.ylabel('Genre')      
plt.show()
```

```python
#Examine the top rated movies that have at least 10 ratings
x=rating_w_names.groupby('movie_title').filter(lambda x: len(x) > 100)
x.groupby('movie_title').rating.mean().sort_values(ascending=False).head(10)
```

```python
#Examine the number of movies in the data set per release year
#movie_info['Year'] = movie_info.release_date.str[7:]
#movie_info.Year = pd.to_numeric(movie_info.Year, errors='coerce').fillna(0).astype(np.int64)
viz6 = sns.countplot(x='Year', data=movie_info[movie_info.Year>1980])
viz6.set_xticklabels(viz6.get_xticklabels(), rotation=90)
```

```python
#Did users from a certain zip code give higher or lower ratings than others?
users_ratings = pd.read_table('u.data', sep='\t', header=None, names=['user_id', 'rating', ], usecols=[0, 2])
users2 = pd.read_table('u.user', sep='|', header=None, names=['user_id', 'zipcode'], usecols=[0, 4])

user_ratings = pd.merge(users_ratings, users2)
user_ratings.groupby('user_id').rating.mean().sort_values(ascending=False)

#Every reviewer had atleast 20 reviews
user_ratings.user_id.value_counts().sort_values()

user_ratings['zipcode_sub'] = user_ratings.zipcode.str[:2]
user_ratings.groupby('zipcode_sub').rating.mean().sort_values()

user_ratings[user_ratings.zipcode_sub.str.isdigit()].groupby('zipcode_sub').rating.mean().sort_values()
#user_ratings[user_ratings.zipcode_sub.str.isdigit()].zipcode_sub.value_counts().sort_values()

#The toughest critics are from Maryland and Kentucky!
#The most lenient critics are from West Virginia and South Dakota!
```
